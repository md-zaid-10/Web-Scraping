```python
import requests
from bs4 import BeautifulSoup
```

```python
# Fetch Data and Store it

page_count = 1

while True:
    URL = f"https://quotes.toscrape.com/page/{page_count}/"
    res = requests.get(URL)

    soup = BeautifulSoup(res.text, "lxml")
    quotes = soup.select("div.quote")

    if not quotes:
        print("No valid pages anymore...")
        break

    with open(f"quotes{page_count}.html", "w") as f:
        f.write(res.text)
        print(f"Downloaded Data from page{page_count}")
    
    page_count = page_count + 1
```

```python
# Extract useful information for page 1

with open("quotes1.html", "r") as f:
    html_content = f.read()

soup = BeautifulSoup(html_content, "lxml")
```

```python
all_quotes = soup.select("div.quote")
life_quotes = []

for q in all_quotes:
    all_tags = []

    for tag in q.select(".tags .tag"):
        all_tags.append(tag.get_text())

    if "life" in all_tags:
        text = q.select_one("span.text").get_text()
        author = q.select_one("small.author").get_text()
        life_quotes.append([text, author])

print(life_quotes)
```

```python
# Extracting for all pages

life_quotes = []

for page in range(1, page_count):

    with open(f"quotes{page}.html", "r") as f:
        html_content = f.read()

    soup = BeautifulSoup(html_content, "lxml")

    all_quotes = soup.select("div.quote")

    for q in all_quotes:
        all_tags = []

        for tag in q.select(".tags .tag"):
            all_tags.append(tag.get_text())

        if "life" in all_tags:
            text = q.select_one("span.text").get_text()
            author = q.select_one("small.author").get_text()
            life_quotes.append([text, author])

print(life_quotes)
```
