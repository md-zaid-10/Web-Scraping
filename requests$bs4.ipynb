```python
from bs4 import BeautifulSoup
```

```python
with open("scraped_data/data1.html", "r") as f:
    html_content = f.read()

soup = BeautifulSoup(html_content, "lxml")
```

```python
soup.find("h1")
all_h3 = soup.find_all("h3")
```

```python
all_h3 = soup.find_all("h3")
#for h3 in all_h3:
#   print(h3)

for h3 in all_h3:
    name = h3.get_text(strip=True)
    print(h3.find_parent("div")["class"])
```

```python
all_countries = []

for h3 in all_h3:

    name = h3.get_text(strip=True)
    print(h3.find_parent("div")["class"])

    population = print(h3.find_next("div").select("span.country-population")[0].get_text(strip=True))

    population = (h3.find_next("div").select_one("span.country-population").get_text(strip=True))

    #print(h3.attrs)

    all_countries.append([name, population])
```

```python
import pandas as pd

df = pd.DataFrame(all_countries, columns=["Name", "Population"])
```

```python
df.to_csv("cleaned_data/data.csv", index=False)
```
